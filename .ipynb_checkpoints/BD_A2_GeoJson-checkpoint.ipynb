# Criando a conexão

## Criando a conexão com o pymysql

```python
p = getpass.getpass()
connection = pymysql.connect(host='localhost', port=3306, user='root', passwd=p) #, db='mysql')
cursor = connection.cursor()
connection.autocommit(True)
```

```python
cursor.execute("show databases;")
for r in cursor.fetchall():
    print(r)
```

## Conectando com o pandas (read_sql)

```python
df_mysql = psql.read_sql('show databases;', con=connection)
df_mysql
```

```python
cursor.execute("DROP DATABASE IF EXISTS educacao_sao_paulo;")
```

```python
cursor.execute("create database educacao_sao_paulo;")
```

## Criando a conexão com o sqlalchemy



```python
#import MySQLdb
from sqlalchemy import create_engine
conectar = f'mysql+pymysql://root:{p}@localhost/educacao_sao_paulo'
engine = create_engine(conectar)
```

```python
connection_engine = engine.connect()
```

# Criando o banco de dados

```python
cursor.execute("use educacao_sao_paulo;")
cursor.execute("""CREATE TABLE geometria (
    id int NOT NULL,
    name varchar(300) NOT NULL,
    description varchar(1000) NOT NULL,
    geometry POLYGON NOT NULL,
    CONSTRAINT geometria_pk PRIMARY KEY (id)
);""")
```

```python
df_mysql = psql.read_sql('show tables;', con=connection)
df_mysql
```

```python
df_mysql = psql.read_sql('SELECT * FROM geometria;', con=connection)
df_mysql
```
